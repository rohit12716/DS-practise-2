```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix , classification_report
import pandas as pd 
from sklearn.model_selection import train_test_split
import seaborn as sns 
import matplotlib.pyplot as plt 



```

```python
data = load_digits() 
```

```python
df = pd. DataFrame(data['data'],
                   columns= data ['feature_names'])

df['target'] = data ['target']
df
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(data['data'],
                 data['target'],
                 random_state=42,
                 test_size=0.2)


model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
 model.score(X_test,y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True,fmt = '.2g')
plt.show()
```

```python
import time
max_score = 0 
all_score = []
all_k = []

for i in range(1,16):
    X_train, X_test, y_train, y_test = \
    train_test_split(data['data'],
                 data['target'],
                 random_state=67,
                 test_size=0.2)


    model = KNeighborsClassifier(n_neighbors=i)
    
    
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    
    
    if score>max_score:
        max_score = score
        
        
        print(f'value of K is: {i} and score is {score}')
        display(clear=True)
    all_score.append(score)
    all_k.append(i)
    time.sleep(1)
    
```

```python
import numpy as np 
k_index = np.argmax(all_score)
```

```python
plt.plot(range(1,16),all_score)

k_index = np.argmax(all_score)

plt.annotate(text = f'Max score: {round(max(all_score),2)},k is {k_index+1}',
            xy = (k_index+1,max(all_score)))

plt.show()
```
